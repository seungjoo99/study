```python
import pandas as pd

train=pd.read_csv('train.csv')
train.head(10)
```

## Age 처리

#### 빈칸 평균으로 메꾸기

```python
train['Age']=train['Age'].fillna(train['Age'].mean())
train.head(10)
```

#### 10대, 20대, 30대, 40대, 50대 나누기

```python
train.loc[train['Age']<10, 'Age']=0
train.loc[(train['Age']>=10) & (train['Age']<20), 'Age']=1
train.loc[(train['Age']>=20) & (train['Age']<30), 'Age']=2
train.loc[(train['Age']>=30) & (train['Age']<40), 'Age']=3
train.loc[(train['Age']>=40) & (train['Age']<50), 'Age']=4
train.loc[train['Age']>=50, 'Age']=5
```

```python
train.head(10)
```

## family size로 합치기

```python
train['Familysize']=train['SibSp']+train['Parch']
train.head()
```

## embarked 채우기

```python
train.isnull().sum() #빈칸 개수 확인
```

```python
train['Embarked'].value_counts() #S가 젤 많구나!
```

```python
train['Embarked']=train['Embarked'].fillna('S')
```

```python
train.isnull().sum()
```

## 성별 처리

```python
train.loc[train['Sex']=='female', 'Sex']=1
train.loc[train['Sex']=='male', 'Sex']=0
```

## 중요한 특성들만 남기기

```python
train=train[['Survived','Sex','Age','Familysize','Embarked']]  #열이 여러개라서 이중 리스트
train.head()
```

## trainset - x , y 나누기

```python
x_train=train[['Sex','Age','Familysize','Embarked']]
y_train=train['Survived']
```
